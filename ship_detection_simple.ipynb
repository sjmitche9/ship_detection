```python
# import required modules
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
f = open('C:/Users/sjmit/practice/lighthouse/w7/d5-day_5/archive/shipsnet.json')
dataset = json.load(f)
f.close()
```

```python
# convert data to numpy arrays
input_data = np.array(dataset['data'])
output_data = np.array(dataset['labels'])
```

```python
# check the shape of the arrays
input_data.shape
```

```python
# reshape data for the model, we now have square images with three layers each (RGB)
channels = 3     #color Channel(RGB)
width = 80
height = 80
X = input_data.reshape(-1, channels, width, height)
```

```python
# set printing options and check new shape
np.set_printoptions(threshold=np.inf)
X.shape
```

```python
# transpose the arrays to the correct order for the model
X = X.transpose([0,2,3,1])
```

```python
# transform our labels/targets to vectors for the model to interpret
from keras.utils import np_utils
y = np_utils.to_categorical(output_data)
```

```python
# plot the first nine images
for i in range(9):
	plt.subplot(330 + 1 + i)
	plt.imshow(X[i])
plt.show()
```

```python
# split the data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7)
```

```python
# normalize and scale the images to values between 0 and 1
X_train = X_train / 255
X_test = X_test / 255
```

```python
# initialize an instance of keras sequential
classifier = Sequential()

# add a convolution layer to look for edges
classifier.add(Convolution2D(32, 3, 3, input_shape = (80, 80, 3), activation = 'relu'))
# add a pooling layer to reduce the size of the data and help prevent over-fitting
classifier.add(MaxPooling2D(pool_size = (2, 2)))
# flatten the image for the dense layers, square image becomes a one dimensional vector
classifier.add(Flatten())
# add two dense layers, sigmoid activation for output between 0 and 1
classifier.add(Dense(512, activation = 'relu'))
classifier.add(Dense(2, activation = 'sigmoid'))

# compile and optimize the model
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# fit the model to our training data
classifier.fit(X_train, y_train, epochs=10, batch_size=200)
```

```python
# print the accuracy and loss
test_loss, test_acc = classifier.evaluate(X_test, y_test)
print('test_acc:', test_acc, 'test_loss', test_loss)
```
